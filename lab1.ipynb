```python
import cairo
import numpy
 
# A simple function to display an image in an ipython notebook
def nbimage( data ):
    from IPython.display import display, Image
    from PIL.Image import fromarray
    from StringIO import StringIO
 
    s = StringIO()
    fromarray( data ).save( s, 'png' )
    display( Image( s.getvalue() ) )
 
WIDTH = 512
HEIGHT = 512
```

```python
import random
```

```python
# for part 1: create an image with 50 random elements

# this is a numpy buffer to hold the image data
img = numpy.zeros( (HEIGHT,WIDTH,4), dtype=numpy.uint8 )
 
# this creates a cairo context based on the numpy buffer
ims = cairo.ImageSurface.create_for_data( img, cairo.FORMAT_ARGB32, WIDTH, HEIGHT )
cr = cairo.Context( ims )

cr.set_line_width( 2.0 )

for i in range(100):
    cr.set_source_rgba( random.random(), 
                       random.random(),
                       random.random(), 
                       1.0 )
    
    cr.move_to( random.random() * 512.0, random.random() * 512.0 )
    cr.line_to( random.random() * 512.0, random.random() * 512.0 )
    cr.stroke()
    cr.move_to( random.random() * 512.0, random.random() * 512.0 )
    cr.curve_to( random.random() * 512.0, random.random() * 512.0, random.random() * 512.0, 
                random.random() * 512.0, random.random() * 512.0, random.random() * 512.0 )
    cr.stroke()

# display the image
nbimage( img )
```
